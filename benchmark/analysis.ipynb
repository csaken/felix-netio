```python
import pandas as pd    

df = pd.io.api.read_csv('results_asyncmsg_vs_libevent.csv')
df["Speedup"] = df["Throughput libevent [MB/s]"]/df["Throughput asyncmsg [MB/s]"]
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(14,10))
plt.plot(df["Size"], df["Throughput libevent [MB/s]"], 'o-', lw=1.6, label="libevent", alpha=0.6, markeredgewidth=0)
plt.plot(df["Size"], df["Throughput asyncmsg [MB/s]"], 'o-', lw=1.6, label="asyncmsg", alpha=0.6, markeredgewidth=0)
plt.grid()
plt.xscale("log", base=2)
plt.legend(loc='lower right')
#plt.xticks(df["Size"])
plt.xlabel("Message Size [Byte]")
_ = plt.ylabel("Throughput [MB/s]")
plt.savefig('throughput.png')
```

```python
plt.figure(figsize=(14,10))
plt.plot(df["Size"], df["Speedup"], 'ro-', lw=1.6, label="libevent", alpha=0.6, markeredgewidth=0)
plt.grid()
plt.xscale("log", base=2)
#plt.xticks(df["Size"])
plt.xlabel("Message Size [Byte]")
_ = plt.ylabel("Speedup")
plt.savefig('speedup.png')
```
